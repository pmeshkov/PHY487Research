```python
import pylab
import numpy as np
from scipy.integrate import solve_ivp
```

```python
def rhs(t, y):
    a = 0.7
    c = 0.007
    b = 1
    p1 = y[0]
    p2 = y[1]
    
    dp1dt = p1 * (a - c * p2)
    dp2dt = p2 * (c*p1 - b)
    
    return [dp1dt, dp2dt]

p0 = [70, 50] # initial condition vector
t0 = 0
tfinal = 30
ts = np.linspace(t0, tfinal, 200)

sol = solve_ivp(rhs, [t0, tfinal], p0, t_eval=ts)

p1 = sol.y[0]
p2 = sol.y[1]

pylab.plot(sol.t, p1, label='rabbits')
pylab.plot(sol.t, p2, '-og', label='foxes')
pylab.legend()
pylab.xlabel('t')
```
